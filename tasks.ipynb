# Handling Missing Value

Missing values are identified and handled to improve data quality and consistency.

```python
df.isnull().sum()
```

```python
df = df.drop(columns=[
    'videos',
    'variations',
    'seller_information'
])
```

```python
df['discount'] = df['discount'].fillna(0)

df['seller_name'] = df['seller_name'].fillna('Unknown')

df['what_customers_said'] = df['what_customers_said'].fillna('No Review')
```

```python
df.isnull().sum()
```

# Creating derived columns

```python
df[['initial_price', 'final_price']].head()
```

```python
df['final_price'] = df['final_price'].str.replace('ГўвЂљВ№', '')

df['final_price'] = df['final_price'].str.replace(',', '')

df['final_price'] = df['final_price'].str.replace('"', '')

df['final_price'] = df['final_price'].astype(float)
```

```python
df['price_difference'] = df['initial_price'] - df['final_price']
```
